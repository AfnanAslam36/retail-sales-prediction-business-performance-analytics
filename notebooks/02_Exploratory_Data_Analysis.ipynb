# Exploratory Data Analysis (EDA)

## Objectives
- Understand sales patterns.
- Analyze the impact of promotions and holidays.
- Compare store performances.
- Identify trends and seasonality.
- Generate business insights for forecasting.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('default')
```

```python
df = pd.read_csv("C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/merged_data.csv", parse_dates=['Date'])

df.head()
```

```python
print("Shape :", df.shape)

display(df.head())

display(df.describe())

df.info()
```

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing = missing[missing > 0]

display(missing)
```

Sales Distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['Sales'],
    bins=50,
    kde=True
)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

Daily Sales Trend

```python
daily_sales = (
    df.groupby('Date')['Sales']
      .sum()
)

plt.figure(figsize=(15,5))

plt.plot(daily_sales)

plt.title("Daily Sales Trend")

plt.xlabel("Date")
plt.ylabel("Total Sales")

plt.show()
```

Monthly Sales Trend

```python
monthly_sales = (
    df.groupby(df['Date'].dt.to_period('M'))['Sales']
      .sum()
)

monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(15,5))

plt.plot(monthly_sales)

plt.xticks(rotation=90)

plt.title("Monthly Sales Trend")

plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

Sales by Month

```python
monthly_avg = (
    df.groupby(df['Date'].dt.month)['Sales']
      .mean()
)

month_names = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(10,5))

plt.bar(month_names, monthly_avg)

plt.title("Average Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales")

plt.show()
```

Sales by Year

```python
yearly_sales = (
    df.groupby(df['Date'].dt.year)['Sales']
      .sum()
)

plt.figure(figsize=(6,5))

plt.bar(
    yearly_sales.index.astype(str),
    yearly_sales.values
)

plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales")

plt.show()
```

Sales by Day of Week

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='DayOfWeek',
    y='Sales',
    estimator=np.mean
)

plt.title("Average Sales by Day of Week")

plt.show()
```

Promo Analysis

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x='Promo',
    y='Sales',
    estimator=np.mean
)

plt.title("Promotion Effect on Sales")

plt.show()
```

State Holiday

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='StateHoliday',
    y='Sales',
    estimator=np.mean
)

plt.title("Average Sales During State Holidays")

plt.show()
```

School Holiday

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x='SchoolHoliday',
    y='Sales',
    estimator=np.mean
)

plt.title("School Holiday Effect")

plt.show()
```

Store Type

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='StoreType',
    y='Sales',
    estimator=np.mean
)

plt.title("Average Sales by Store Type")

plt.show()
```

Assortment

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x='Assortment',
    y='Sales',
    estimator=np.mean
)

plt.title("Average Sales by Assortment")

plt.show()
```

Competition Distance

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    data=df,
    x='CompetitionDistance',
    y='Sales',
    alpha=0.4
)

plt.title("Competition Distance vs Sales")

plt.show()
```

Open vs Closed Stores

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x='Open',
    y='Sales',
    estimator=np.mean
)

plt.title("Open vs Closed Store Sales")

plt.show()
```

Top 10 Stores

```python
top10 = (
    df.groupby('Store')['Sales']
      .sum()
      .nlargest(10)
)

plt.figure(figsize=(12,5))

top10.plot(kind='bar')

plt.title("Top 10 Stores by Sales")

plt.ylabel("Sales")

plt.show()
```

Bottom 10 Stores

```python
bottom10 = (
    df.groupby('Store')['Sales']
      .sum()
      .nsmallest(10)
)

plt.figure(figsize=(12,5))

bottom10.plot(kind='bar')

plt.title("Bottom 10 Stores by Sales")

plt.ylabel("Sales")

plt.show()
```

Boxplots for Outlier Detection

```python
plt.figure(figsize=(12,5))

sns.boxplot(
    x=df['Sales']
)

plt.title("Sales Outlier Detection")

plt.show()
```

Sales vs Customers

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='Customers',
    y='Sales',
    alpha=0.4
)

plt.title("Customers vs Sales")

plt.show()
```

Pairplot

```python
pair_df = df[
    [
        'Sales',
        'Customers',
        'CompetitionDistance',
        'Promo'
    ]
]

sns.pairplot(pair_df)

plt.show()
```

Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(12,10))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Matrix")

plt.show()
```
