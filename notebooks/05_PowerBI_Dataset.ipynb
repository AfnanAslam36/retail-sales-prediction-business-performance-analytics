```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/featured_data.csv",
    parse_dates=["Date"]
)

df.head()
```

```python
print(df.shape)

display(df.head())

df.info()
```

```python
columns_to_remove = [
    "Lag_1",
    "Lag_7",
    "Lag_30",
    "Rolling7Mean",
    "Rolling30Mean",
    "Rolling7Std"
]

df.drop(
    columns=columns_to_remove,
    inplace=True
)
```

```python
df.rename(columns={

    "Promo":"Promotion",

    "Promo2":"Continuous Promotion",

    "CompetitionDistance":"Competition Distance",

    "CompetitionMonthsOpen":"Competition Months Open",

    "PromoMonths":"Promotion Duration (Months)",

    "CompetitionNearby":"Competition Nearby",

    "StoreType":"Store Type",

    "StateHoliday":"State Holiday"

}, inplace=True)
```

```python
df["Promotion"] = df["Promotion"].replace({

    0:"No",

    1:"Yes"

})

df["Continuous Promotion"] = df["Continuous Promotion"].replace({

    0:"No",

    1:"Yes"

})

df["Competition Nearby"] = df["Competition Nearby"].replace({

    0:"No",

    1:"Yes"

})

df["IsHoliday"] = df["IsHoliday"].replace({

    0:"No",

    1:"Yes"

})

df["IsWeekend"] = df["IsWeekend"].replace({

    0:"Weekday",

    1:"Weekend"

})
```

```python
df.drop(columns=["SalesPerCustomer"], inplace=True)
```

```python
df.drop(columns=["Weekday"], inplace=True)
```

```python
df.drop(columns=["MonthName"], inplace=True)
```

```python
print(df.shape)

display(df.head())

df.info()
```

```python
df.to_csv(

    "C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/powerbi_dataset.csv",

    index=False

)

print("Power BI Dataset Exported Successfully!")
```
