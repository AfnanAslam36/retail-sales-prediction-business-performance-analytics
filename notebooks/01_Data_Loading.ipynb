# Retail Sales Prediction & Business Performance Analytics

## Notebook 01 – Data Loading & Understanding

### Objectives
- Load the Rossmann datasets.
- Understand each feature.
- Check data quality.
- Handle missing values in the store dataset.
- Merge datasets.
- Save the merged dataset for further analysis.

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/train.csv", low_memory=False)
store = pd.read_csv("C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/store.csv")
```

```python
print("=" * 60)
print("TRAIN DATASET")
print("=" * 60)

print("\nShape")
print(train.shape)

print("\nFirst Five Rows")
display(train.head())

print("\nData Types")
display(train.dtypes)

print("\nInformation")
train.info()

print("\nStatistical Summary")
display(train.describe(include='all'))
```

```python
print("=" * 60)
print("STORE DATASET")
print("=" * 60)

print("\nShape")
print(store.shape)

print("\nFirst Five Rows")
display(store.head())

print("\nData Types")
display(store.dtypes)

print("\nInformation")
store.info()

print("\nStatistical Summary")
display(store.describe(include='all'))
```

```python
print("Train Dataset Missing Values")

missing_train = (train.isnull().sum().sort_values(ascending=False))

display(missing_train[missing_train > 0])

print("\nStore Dataset Missing Values")

missing_store = (store.isnull().sum().sort_values(ascending=False))

display(missing_store[missing_store > 0])
```

```python
store['PromoInterval'].value_counts()
```

```python
store['CompetitionDistance'] = store['CompetitionDistance'].fillna(
    store['CompetitionDistance'].median()
)

store['CompetitionOpenSinceMonth'] = store['CompetitionOpenSinceMonth'].fillna(0)

store['CompetitionOpenSinceYear'] = store['CompetitionOpenSinceYear'].fillna(0)

store['Promo2SinceWeek'] = store['Promo2SinceWeek'].fillna(0)

store['Promo2SinceYear'] = store['Promo2SinceYear'].fillna(0)

store['PromoInterval'] = store['PromoInterval'].fillna('No_Promo')
```

```python
store.isnull().sum()
```

```python
print("Train Duplicates :", train.duplicated().sum())
print("Store Duplicates :", store.duplicated().sum())
```

```python
train['Date'] = pd.to_datetime(train['Date'])

print(train['Date'].head())
print(train['Date'].dtype)
```

```python
df = train.merge(
    store,
    on='Store',
    how='left'
)

print("Merged Dataset Shape:", df.shape)

display(df.head())
```

```python
print(df.info())

print("\nMissing Values After Merge")

display(
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)
```

```python
df.to_csv("C:/Projects/Retail Sales Prediction & Business Performance Analytics/data/merged_data.csv", index=False)

print("Merged dataset saved successfully!")
```
